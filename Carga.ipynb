```python
import pandas as pd
import numpy as np
```

```python
sucursal = pd.read_csv('Datasets\sucursal.csv')
```

```python
sucursal.info()
```

```python
# Primary key sin duplicados
sucursal[sucursal['id'].duplicated()]
```

```python
# Lista de sucursalId
sucursalId = list()
for ind, elem in enumerate(sucursal['id']):
    n= elem.split('-')
    sucursalId.append(int(n[2]))

# Agregamos la nueva columna
sucursal['sucursalId'] = sucursalId
```

```python
# Vemos que no hay duplicados
sucursal[sucursal.duplicated()]
```

```python
sucursal= sucursal.rename(columns={'id': 'principalId'})
```

```python
# ERROR: 1262 
# en localidad hay una coma
sucursal[sucursal['principalId'] == '24-1-194']
```

```python
cambio = list()
for ind, elem in enumerate(sucursal['localidad'] ):
    elem= elem.replace(',','/')
    cambio.append(elem)
sucursal['Localidad'] = cambio
sucursal = sucursal.drop('localidad', axis=1)
```

```python
# En direccion tambien hay comas
sucursal[sucursal['principalId'] == '9-1-655']
```

```python
cambio = list()
for ind, elem in enumerate(sucursal['direccion'] ):
    elem= elem.replace(',','/')
    cambio.append(elem)
sucursal['Direccion'] = cambio
sucursal = sucursal.drop('direccion', axis=1)
```

```python
# En sucursalNombre tambien hay comas
sucursal[sucursal['principalId'] == '9-1-731']
```

```python
cambio = list()
for ind, elem in enumerate(sucursal['sucursalNombre'] ):
    elem= elem.replace(',','/')
    cambio.append(elem)
sucursal['SucursalNombre'] = cambio
sucursal = sucursal.drop('sucursalNombre', axis=1)
```

```python
# Reordenar columnas
sucursal=sucursal.reindex(columns=['principalId', 'comercioId','banderaId', 'sucursalId', 'banderaDescripcion', 'comercioRazonSocial', 
                                    'provincia', 'Localidad', 'Direccion', 'lat', 'lng','SucursalNombre', 'sucursalTipo'])
```

```python
sucursal.describe()
```

```python
sucursal.to_csv('tabla_sucursal.csv', index= False)
sucursal.to_csv('C:\\ProgramData\\MySQL\\MySQL Server 8.0\\Uploads\\sucursal.csv', index= False)
sucursal.head(1)
```

```python
producto = pd.read_parquet('Datasets\producto.parquet')
producto.head()
```

```python
producto = producto.fillna('sin dato')
```

```python
producto= producto.rename(columns={'id': 'productoId'})
```

```python
# Quitamos columnas
producto = producto.drop(['categoria1','categoria2','categoria3'], axis=1)
```

```python
# No hay duplicados
producto[producto['productoId'].duplicated()]
```

```python
# Vemos una coma en nombre
producto[producto['productoId'] == '0000077961709']
```

```python
# Nueva columna con Marca sin ceros adelantes
cambio = list()
for ind, elem in enumerate(producto['nombre'] ):
    elem = elem.replace(',', ' ')
    cambio.append(elem)
producto['Nombre'] = cambio
producto = producto.drop('nombre', axis=1)
```

```python
producto = producto.reindex(columns=['productoId', 'marca', 'Nombre', 'presentacion'])
```
